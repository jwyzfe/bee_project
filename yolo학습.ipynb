```python
import yaml
# YOLOv8 학습과 검증에 사용된 train, valid date 가 저장되어 있는 디렉토리 경로
data = { 'train': '/content/drive/MyDrive/fruits/yolov8/data/train',
        'val': '/content/drive/MyDrive/fruits/yolov8/data/val',
        'test': '/content/drive/MyDrive/fruits/yolov8/data/test',
        'name' : ['disease_apple'],
        'nc': 1}

# 데이터 경로와 클래스 정보를 저장하고 있는 딕셔너리 객체 data를 YOLOv8 학습에 필요한 질병_사과.yaml 으로 저장
with open('/content/drive/MyDrive/fruits/yolov8/data/질병_사과.yaml', 'w') as f:
    yaml.dump(data, f)

# 질병_사과_yaml 읽어서 화면에 출력
with open('/content/drive/MyDrive/fruits/yolov8/data/질병_사과.yaml', 'r') as f:
    질병_사과_yaml = yaml.safe_load(f)
    display(질병_사과_yaml)
```
